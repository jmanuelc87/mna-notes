# Morphological Operations

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Dilation Operation

```python
image = cv2.imread('../datasets/esferas.png')

shape = (5, 5)
rect = cv2.getStructuringElement(cv2.MORPH_RECT, shape)
open = cv2.getStructuringElement(cv2.MORPH_CROSS, shape)
ellipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, shape)

morph1 = cv2.dilate(image, rect, iterations=1)
dilation2 = cv2.dilate(image, open, iterations=1)
dilation3 = cv2.dilate(image, ellipse, iterations=1)

plt.figure(figsize=(15,12))

plt.subplot(1,3,1)
plt.imshow(morph1[:,:,::-1])

plt.subplot(1,3,2)
plt.imshow(dilation2[:,:,::-1])

plt.subplot(1,3,3)
plt.imshow(dilation3[:,:,::-1])
```

### Erosion Operation

```python
image = cv2.imread('../datasets/esferas.png')

shape = (5, 5)
rect = cv2.getStructuringElement(cv2.MORPH_RECT, shape)
open = cv2.getStructuringElement(cv2.MORPH_CROSS, shape)
ellipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, shape)

erode1 = cv2.erode(image, rect, iterations=1)
erode2 = cv2.erode(image, open, iterations=1)
erode3 = cv2.erode(image, ellipse, iterations=1)

plt.figure(figsize=(15,12))

plt.subplot(1,3,1)
plt.imshow(erode1[:,:,::-1])

plt.subplot(1,3,2)
plt.imshow(erode2[:,:,::-1])

plt.subplot(1,3,3)
plt.imshow(erode3[:,:,::-1])
```

### Opening operation

```python
image = cv2.imread('../datasets/esferas.png')

shape = (5, 5)
rect = cv2.getStructuringElement(cv2.MORPH_RECT, shape)
open = cv2.getStructuringElement(cv2.MORPH_CROSS, shape)
ellipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, shape)

morph1 = cv2.morphologyEx(image, cv2.MORPH_OPEN, rect)
morph2 = cv2.morphologyEx(image, cv2.MORPH_OPEN, open)
morph3 = cv2.morphologyEx(image, cv2.MORPH_OPEN, ellipse)

plt.figure(figsize=(15,12))

plt.subplot(1,3,1)
plt.imshow(morph1[:,:,::-1])

plt.subplot(1,3,2)
plt.imshow(morph2[:,:,::-1])

plt.subplot(1,3,3)
plt.imshow(morph3[:,:,::-1])
```

### Top hat operation

```python
image = cv2.imread('../datasets/esferas.png')

shape = (5, 5)
rect = cv2.getStructuringElement(cv2.MORPH_RECT, shape)
open = cv2.getStructuringElement(cv2.MORPH_CROSS, shape)
ellipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, shape)


# A top hat operation is defined as the difference between the input image and the opening of the image.
morph1 = cv2.morphologyEx(image, cv2.MORPH_TOPHAT, rect)
morph2 = cv2.morphologyEx(image, cv2.MORPH_TOPHAT, open)
morph3 = cv2.morphologyEx(image, cv2.MORPH_TOPHAT, ellipse)

plt.figure(figsize=(15,12))

plt.subplot(1,3,1)
plt.imshow(morph1[:,:,::-1])

plt.subplot(1,3,2)
plt.imshow(morph2[:,:,::-1])

plt.subplot(1,3,3)
plt.imshow(morph3[:,:,::-1])
```
