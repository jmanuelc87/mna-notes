# Histogram Equalization

```python
import os
import cv2
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Grayscale histograms without a mask

```python
image = cv2.imread('../datasets/image4.bmp')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([image_gray], [0], None, [256], [0, 256])

sns.lineplot(hist)
```

```python
image = cv2.imread('../datasets/mountain.jpg')

colors = ('b', 'g', 'r')

for i, color in enumerate(colors):
    hist = cv2.calcHist([image], [i], None, [256], [0,256])
    plt.plot(hist, color=color)
plt.title("Image Histogram GFG")
```

### Histogram equalization

```python
image = cv2.imread('../datasets/mountain.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_gray_eq = cv2.equalizeHist(image_gray)

hist = cv2.calcHist([image_gray_eq], [0], None, [256], [0,256])
hist1 = cv2.calcHist([image_gray], [0], None, [256], [0,256])


fig, ax = plt.subplots(2,2)
ax = ax.ravel()
ax[0].imshow(image_gray_eq, cmap='gray')
ax[1].plot(hist)
ax[2].imshow(image_gray, cmap='gray')
ax[3].plot(hist1)

fig.tight_layout(pad=1.7)
```

### Contrast Limited Adaptive Histogram Equalization

```python
clahe = cv2.createCLAHE(clipLimit=2.0)
gray_image_clahe = clahe.apply(image_gray)


hist = cv2.calcHist([gray_image_clahe], [0], None, [256], [0,256])


fig, ax = plt.subplots(2,2)
ax = ax.ravel()
ax[0].imshow(gray_image_clahe, cmap='gray')
ax[1].plot(hist)
ax[2].imshow(image_gray, cmap='gray')
ax[3].plot(hist1)

fig.tight_layout(pad=1.7)
```
